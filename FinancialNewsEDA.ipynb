```python
import pandas as pd
import numpy as np
```

```python
cnbc_df = pd.read_csv("/content/cnbc_headlines.csv")
guardian_df = pd.read_csv("/content/guardian_headlines.csv")
reuters_df = pd.read_csv("/content/reuters_headlines.csv")

cnbc_df.head()
```

```python
reuters_df.head()
```

```python
guardian_df.head()
```

```python
cnbc_df = cnbc_df.dropna()
guardian_df = guardian_df.dropna()
reuters_df = reuters_df.dropna()
```

```python
df = pd.concat([cnbc_df, guardian_df, reuters_df], ignore_index=True)
df.head()
```

```python
df.info()
```

```python
df.dropna()
```

```python
from dateutil import parser

def time_parser(t):
  try:
    return parser.parse(t, fuzzy=True)
  except:
    return pd.NaT

df["New_Time"] = df['Time'].apply(time_parser)
df.head()
```
